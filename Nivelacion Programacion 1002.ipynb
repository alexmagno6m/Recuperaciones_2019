# Recuperacion General de Programacion - Grado 1001
## Instrucciones
> 1. Si usted perdio dos periodos o menos debe realizar unicamente la recuperacion de estos periodos.
> 2. Si perdio tres o mas periodos debe realizar la recuperacion total de la materia (los 4 periodos)
> 3. Las recuperacion consiste en un trabajo escrito y un conjunto de ejercicios practicos que se le entregaran el dia que sustente el trabajo escrito.
>> El trabajo escrito debe estar hecho a mano y en hojas cuadriculadas, no se reciben cuadernos.
> 4. Tanto el trabajo como el ejercicio deben ser sustentados verbalmente y de manera practica.
> 5. La nota de la recuperacion sera maximo de 3.5 (Basico en la escala general del colegio)
